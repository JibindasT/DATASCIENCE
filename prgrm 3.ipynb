```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(12,12))
font1 = {'color':'blue'}
font2 = {'color':'orange'}
plt.suptitle("GRAPH",fontsize = 30)
#plot 1
xpoints = np.array([1, 6])
ypoints = np.array([3,5])

plt.subplot(3,3,1)
plt.xlabel("x-axis",fontdict = font1)
plt.ylabel("y-axis",fontdict = font1)
plt.plot(xpoints,ypoints,'o')





#plot 2
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6,2,8,1])
plt.subplot(3,3,2)
plt.plot(y1,'*-',ms=15,linewidth='5',color='orange')
plt.plot(y2,linewidth='1',linestyle='--',color='blue')

#plot 3
ypoints = np.array([3, 8, 1, 10, 5, 7])
plt.subplot(3,3,3)
plt.plot(ypoints,marker ='o',mfc='yellow',mec='black')

#plot 4

x = np.array([80, 85, 90, 95, 100])
y = np.array([10,20,30,40,50])


plt.subplot(3,3,4)
plt.plot(x, y)

plt.grid()



#plot 5


x = np.array([5,10,5,20,22,24,26,25])
y = np.array([20,30,40,42,43,44,50,60])
sizes = np.array([20,50,100,200,500,200,70,80])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple"])
plt.subplot(3,3,5)
plt.scatter(x, y, s=sizes, alpha=0.5,c=colors)


#plot 6
x = np.array(["A", "B", "C"])
y = np.array([ 8, 3, 10])

plt.subplot(3,3,6)
plt.bar(x,y,color='red',width=0.1)


#plot 7
y = np.array([35,20,20,25])
plt.subplot(3,3,7)
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
plt.pie(y,labels=mylabels, startangle = 90 , explode=myexplode,shadow='true')
plt.legend()


#plot 8
x = np.array([10,20,30,40,50,60,70])
y = np.array([5,10,5,20,5,30,5])
plt.subplot(3,3,8)
plt.title("diagram")
plt.xlabel("flow")
plt.ylabel("time")

plt.plot(x,y ,'o--r')
plt.grid()


#plot 9
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6,2,8,1])
plt.subplot(3,3,9)
plt.plot(y1,linewidth='5', linestyle = 'dotted',color='black')
plt.plot(y2,linewidth='1',linestyle='--',color='hotpink')
plt.show()
```
